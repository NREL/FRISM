## Determine zones that are withing bounds, save their GEOID

```python
in_geoids = []

for i, row in zones.iterrows():
    # find the centroid of the zone in question
    geoid = row['GEOID']
    p = zones[zones['GEOID'] == geoid]['geometry'].centroid.item()
    
    # Check if the zone is within bounds
    for b in bounds['geometry']:
        if p.within(b):
            print('I am in')
            in_geoids.append(geoid)
            break
```

```python
in_zones = zones[zones['GEOID'].isin(in_geoids)] # Extracting that are inside the boudns
```

```python
in_zones.to_file('in_zones.shp')    # saving those to shapefile
```

## Mapping outside zones to inside zones

```python
len(zones)
```

```python
mapping = pd.DataFrame(columns=['MESOZONE', 'BoundaryZONE'], index=range(len(zones)-len(in_zones)))
```

```python
for index, row in zones.iterrows():
    curr_id = row['GEOID']
    if curr_id not in in_geoids:
        D = 999999
        Id = None 
        g_Id = None
        p = row['geometry'].centroid
        for i,z in enumerate(in_zones['geometry']):
            distance = p.distance(z)
            ID = in_zones.iloc[i]['MESOZONE']
            GID = in_zones.iloc[i]['GEOID']
            if distance <= D:
                D = distance
                Id = ID
                g_Id = GID
        print('boundary ID for zone ', row['GEOID'], ' is ', Id, ' corres ', g_Id)
        mapping.at[index, 'MESOZONE'] = row['MESOZONE']
        mapping.at[index, 'BoundaryZONE'] = Id
```

```python
mapping = mapping[mapping['MESOZONE'].notna()]
```

```python
mapping.to_csv('External_Zones_Mapping.csv', index=False)
```
