```python
import pandas as pd
import numpy as np
import pyarrow
import geopandas as gpd
import shapely.speedups
from shapely.geometry import Point, Polygon
shapely.speedups.enable()
```

```python
from os import listdir
```

```python
#x,y=(-123.56638,38.81437),(-121,720949, 38.90707) ( 36.917329) 

lon_max=-121.720949
lon_min=-123.56638
lat_max=38.90707
lat_min=36.917329
```

```python
file_dir = '/Users/kjeong/1_Work/1_3_RMobility/INRIX/trips_raw/trips/'
```

```python
df_md=pd.DataFrame()
df_hd=pd.DataFrame()
for file_nm in listdir(file_dir):
    print (file_nm)
    df=pd.read_parquet(file_dir+file_nm, engine='pyarrow')
    df_trip=df[(df.start_lat<=lat_max) & (df.start_lat>=lat_min) 
                 & (df.start_lon<=lon_max) & (df.start_lon>=lon_min)]
    df_md=pd.concat([df_md, df_trip[df_trip.vehicle_weight_class==2]],ignore_index=True)   
    df_hd=pd.concat([df_hd, df_trip[df_trip.vehicle_weight_class==3]],ignore_index=True)
    
```

```python
df_md.to_csv("md_trips.csv")
df_hd.to_csv("hd_trips.csv")
```
